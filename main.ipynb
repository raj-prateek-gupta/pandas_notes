handling Missing Values 


```python
#dropna :droping the NaN value 
# it drops the complete row that contains NaN values 
df=pd.read_csv(r"C:\Users\pk247\OneDrive\Desktop\machine learning\dataset\train.csv")

df.dropna()
```

parameters in dropna 

```python
#axis parameters 
#column wise dropping 
df.dropna(axis=1)
```

```python
#rowise dropping 

df.dropna(axis=0)
```

```python
#how parameter : for removing the entiire row
```

```python
#using "how "parameter for dropping the entire row that contains the NaN values 

#1.using any parameter

df=pd.read_csv(r"C:\Users\pk247\OneDrive\Desktop\machine learning\dataset\train_2.csv")

df.dropna(how="any").head(10)
```

```python
#using all parameter

df.dropna(how="all").head(10)
```

```python
# removing NaN values from perticular columns 
# using subset parameter

df.dropna(subset=["Cabin"]).head(10)
```

```python
df.dropna(inplace=True)

df
```

```python
# dropping the rows according to the count of NaN values 
#using thresh parameter
#parameter tells about the How many NaN values a single contained

df.dropna(thresh=1).head(10)

```
